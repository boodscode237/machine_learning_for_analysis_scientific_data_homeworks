## Выбрать и обосновать метрику для измерения качества.


<b>Предлагаемый показатель: Accuracy with F1-score.
Обоснование:</b>
<hr>
Хотя точность является простым и широко используемым показателем, она может вводить в заблуждение в несбалансированных наборах данных. В этом случае, когда один класс (без сердечно-сосудистых заболеваний) может быть значительно больше, модель может просто постоянно прогнозировать класс большинства и достигать высокой точности, не понимая по-настоящему данные.

Поэтому я предлагаю использовать точность в сочетании с показателем F1. Показатель F1 уравновешивает точность (правильное определение заболеваний сердца) и полноту (не пропускает фактические случаи заболеваний сердца), обеспечивая более надежную оценку потенциально несбалансированных наборов данных. Эта комбинация проливает свет как на способность модели правильно классифицировать оба класса, так и на ее чувствительность к выявлению истинных случаев сердечно-сосудистых заболеваний.

Этот двойной подход устраняет потенциальную предвзятость из-за классового дисбаланса и обеспечивает более полную картину эффективности модели в прогнозировании риска сердечно-сосудистых заболеваний.


***Formulas:***(1, 2)

*Accuracy = (True Positives + True Negatives) / (Total Number of Predictions)*

*Precision = True Positives / (True Positives + False Positives)*

*Recall = True Positives / (True Positives + False Negatives)*

**F1-score:** = 2 * (Precision * Recall) / (Precision + Recall)
